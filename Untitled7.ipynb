```python
import matplotlib.pyplot as plt
import numpy as np
X = np.array([0,1,2,3,4])
Y = np.array([2,4,3,5,0])

plt.plot(X,Y)
plt.show()



data_1 = np.linspace(1,5,100)
data_2 = np.log(data_1)
data_3 = data_1**2
plt.plot(data_1, data_1, 'b--', label='linear')
plt.plot(data_1, data_2, 'g:', label='log')
plt.plot(data_1, data_3, 'r-.', label='quadratic')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Runtime')
plt.show()
```
