```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
Data = pd.read_csv('CompanyABCProfit.csv')
print(Data)
```

It has 200 rows
scale means standard deviation
loc means mean 

```python
plt.bar(Data['Year'],Data["Profit(Rs '000)"])
plt.ylabel("Profit in thousands")
plt.xlabel("year")
plt.title("Company Profits by Year")
```

I would expect an overall normal distribution and maybe trending upwards as the years continue

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(1825,2020,1)
ax.plot(Data['Year'],Data["Profit(Rs '000)"])
ax.set_title("Profit in thousands")
ax.set_xlabel('Year')
ax.set_ylabel('profit')
ax.grid(True)
```

```python
plt.hist(Data["Profit(Rs '000)"])
plt.xlabel("Profit (Rs '000)")
plt.ylabel("Frequency")
plt.title("Histogram of Profit Data")

fig, ax = plt.subplots()
profitSTD = np.std(Data["Profit(Rs '000)"])
x = np.linspace(min(Data["Profit(Rs '000)"]), max(Data["Profit(Rs '000)"]), 100)
p = norm.pdf(x, loc=np.mean(Data["Profit(Rs '000)"]), scale=profitSTD)
ax.plot(x, p, 'r-', lw=2)
plt.plot(x, p, 'r-', lw=2)
plt.show()
```

```python
profitSTD = np.std(Data["Profit(Rs '000)"])
profitMean = np.mean(Data["Profit(Rs '000)"])

x = np.linspace(min(Data["Profit(Rs '000)"]), max(Data["Profit(Rs '000)"]), 100)
p = norm.pdf(x, loc=profitMean, scale=profitSTD)
plt.hist(Data["Profit(Rs '000)"], density=True)

plt.plot(x, p, 'r-', lw=2)

plt.xlabel('Profit (Rs \'000)')
plt.ylabel('Density')
plt.title('Histogram and Probability Density Function of Profit Data')

plt.show()
```
